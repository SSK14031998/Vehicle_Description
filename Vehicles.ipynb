```python
import pandas as pd       
```

```python
data=pd.read_csv('Australian Vehicle Prices.csv')
```

```python
data
```

```python
df = pd.DataFrame(data)
Transmission='Manual'
filter_Data = df[df['Transmission'] == Transmission]
filter_Data 
```

```python
Transmission='Manual'
Condition='NEW'
filter_Data = df[(df['Transmission'] == Transmission) & (df['UsedOrNew'] == Condition)]
filter_Data # Applying multiple filters
```

```python
Transmission='Megane'
res = df['Title'].str.contains(Transmission).fillna(False)
print(df[res])
print(df[res].shape)
```

```python
Drive='Other'
filter_Data = df[df['DriveType']==Drive]
filter_Data
```
